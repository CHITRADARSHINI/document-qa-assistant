```python
%pip install pypdf
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Load your PDF
loader = PyPDFLoader("..\data\Psycology of money.pdf")  # change filename
docs = loader.load()

print(f"Total pages loaded: {len(docs)}")
print(f"\nSample content:\n{docs[0].page_content[:500]}")
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

chunks = splitter.split_documents(docs)

print(f"Total chunks created: {len(chunks)}")
print(f"\nSample chunk:\n{chunks[0].page_content}")
```

```python
import os
os.environ["HF_HUB_DISABLE_SYMLINKS_WARNING"] = "1"
from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="all-MiniLM-L6-v2"  # free, runs locally, no API key needed
)

# Test it
sample_embedding = embeddings.embed_query("test sentence")
print(f"Embedding dimension: {len(sample_embedding)}")
# Should print 384
```

```python
from langchain_community.vectorstores import FAISS

# Build the vector store from your chunks
vectorstore = FAISS.from_documents(chunks, embeddings)

# Save it locally so you don't rebuild every time
vectorstore.save_local("../data/faiss_index")

print("Vector store created and saved!")
```

```python
# Load it back
vectorstore = FAISS.load_local(
    "../data/faiss_index",
    embeddings,
    allow_dangerous_deserialization=True
)

retriever = vectorstore.as_retriever(search_kwargs={"k": 3})

# Ask a question — it should return the 3 most relevant chunks
question = "What is this document about?"  # change to something relevant to your PDF
results = retriever.invoke(question)

for i, doc in enumerate(results):
    print(f"\n--- Chunk {i+1} ---")
    print(doc.page_content)
```

```python
import os
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langchain.chains import RetrievalQA
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings

load_dotenv("../.env")

# Re-load vectorstore and retriever (add these)
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
vectorstore = FAISS.load_local(
    "../data/faiss_index",
    embeddings,
    allow_dangerous_deserialization=True
)
retriever = vectorstore.as_retriever(search_kwargs={"k": 3})

# LLM setup
llm = ChatGroq(
    model_name="llama-3.1-8b-instant",  # updated model name
    api_key=os.getenv("GROQ_API_KEY")
)

qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=retriever,
    return_source_documents=True
)

# Ask your first question!
response = qa_chain.invoke({"query": "What is this document about?"})

print("Answer:", response["result"])
print("\nSources used:")
for doc in response["source_documents"]:
    print("-", doc.page_content[:200])
```
